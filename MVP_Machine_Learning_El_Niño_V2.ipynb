> O que é o El Niño?

O El Niño é um fenômeno natural que surge da interação entre a atmosfera e o oceano na região tropical do oceano Pacífico. É caracterizado pelo aquecimento anormal e persistente da superfície do Oceano Pacífico na região da Linha do Equador, podendo se estender desde a costa da América do Sul até o meio do Pacífico Equatorial.

Durante o El Niño, o movimento dos ventos alíseos no Oceano Pacífico - que normalmente empurra as águas da superfície para o oeste permitindo que as águas mais profundas e frias subam - enfraquecem. Quando os ventos alísios estão enfraquecidos ou invertem a direção, essa troca de águas não ocorre e as águas mais quentes permanecem por mais tempo paradas na superfície, podendo aumentar a sua temperatura em até 3°C ou mais acima da média.

Este fenômeno normalmente começa a se formar no segundo semestre do ano, sendo caracterizado por um aumento de temperatura de 0,5°C  das águas  por um período de no mínimo seis meses. Vale lembrar que o fenômeno não tem um período de duração definido podendo persistir por até dois anos ou mais.

O El Niño é ligado a mudanças significativas na atmosfera, conhecidas como Oscilação Sul (Southern Oscilation - SO). O fenômeno completo, incluindo variações oceâncicas e atmosféricas, é chamado pelos cientistas de El Niño - Southern Oscilation (ENSO) ou El Niño Oscilação Sul (ENOS) em português. Durante o El Niño áreas de pressão atmmosférica maiores do que o normal ocorrem na Australia, Indonésia, Sudeste Asiático, Filipinas, causando condições de seca. Tais condições também são predominantes no Havaí, regiões da África e do Nordeste brasileiro. Já as áreas de baixa pressão ocorrem ao longo da costa oeste e no extremo sul da América do Sul, causando o aumento na quantidade de chuvas. As regiões que costuma enfrentar secas durante o El Niño tendem a se tornar chuvosas durante os eventos de La Niña, que ao contrário do el Niño, é o resfriamento das águas do Oceano Pacífico.

## 1.2 Atributos do Conjunto de dados

O monitoramento do ENOS foca primariamentes nas temperatura superficial da águas (Sea Surface Temperature - SST) em 4 regiões geograficas do Pacífico Equatorial.Existem diversos índices utilizados para monitorar a região tropical do Pacífico, todos baseados na temperatura superficial da águas (SST). O ídice Niño 3.4 e o íncice oceânico Niño (Oceanic Niño Index - ONI ) são os mais comumete usados.

As atributos contidos no conjunto de dados são os seguintes:



*   Date - informação de dia, mês e ano dos dados coletados
*   Year - ano em qe os dados foram coletados
*   Month - mês em que os dados foram coletados
*   Global Temperature Anomalies: é o desvio da tempertura medida em relação ao valor de referência, que neste caso é uma média de longo prazo. Uma anomalia positiva indica que a temperatura observada foi mais quente que o valor de referência, enquanto uma anomalia negativa indica que a temperatura observada foi mais fria que o valor de referência.
*   Nino 1+2 SST (0-10S,90W-80W): Compreende a menor região das  regiões onde o SST é monitorado, corresponde à região próxima a costa da América do Sul onde o fenômeno foinotado pela primeira vez pela população local. Este índice tende a ter a maior variação dos índices SST.
*   Nino 1+2 SST Anomalies: diferença entre a temperatura obsservada e a média histórica para essa região.
*   Niño 3 (5N-5S,150W-90W): Essa região foi inicialmente usada como foco para monitoramento e predição do El Niño, porém com os avanços dos estudos percebeu-se que a região mais importante para o monitoramento do ENOS encontrava-se mais a oeste.
*   Nino 3 SST Anomalies: diferença entre a temperatura obsservada e a média histórica para essa região.
*   Niño 3.4 (5N-5S,170W-120W): As anomalias dessa região são consideradas pelos psquisadores como representantes da média equatorial das SST no Pacífico. Este ídice é calculado com base em uma média móvel de 5 meses e os períodos de El Niño ou La Niña são definidos quando o índice Niño 3.4 excede +/- 0.4 C por um período de seis meses ou mais.
*   Nino 3.4 SST Anomalies: diferença entre a temperatura obsservada e a média histórica para essa região.
*   Niño 4 (5N-5S,160W-150W): Esse ídice captura as anomalias SST da região central do Pacífico equatorial. Essa região tende a ter menos variações do que as outras.
*   Nino 4 SST Anomalies:diferença entre a temperatura obsservada e a média histórica para essa região.
*   TNI - Trans-Niño Idex: É adiferença entre as anomalias SST normalizadas das regiões Niño 1+2 e Niño 4.
*   PNA - Pacific/North American Index: É um índce diário que anomalias de pressão de 500 milibar medido em quatro locais do Oceano Pacífico e América do Norte no hemisfério norte. A fase positiva do PNA é caracterizado por pressão atmosférica acima da média nas proximidadesdo Havaí e no oeste da América do Norte e pressão abaixo da média no Sul do Alasca e no Sudeste dos EUA. Este índice aparece mais positivo em anos com El niño.
*   OLR - Outgoing Longwave Radiation: É observado entre as longitudes 160E a 160 W. OLR negativo indica aumento na convecção e mais cobertura de nuvem tipicas do El Niño. Maior atividade de convecção na parte central e oriental do Pacífico Equatorial implica em maiores e mais frios topos de nuvem que emitem menos radiação infra-vermelha para o espaço, quando o céu está limpo uma grande quantidade de ondas longas é emitidas para a atmosfera. Em céu nublado a radiação fica retidas e não escapa.
*   SOI - Southern Oscilaton Index: Caracteriza as variações da pressão atmosferica medida ao nível do mar entre a parte ocidental e oriental do Pacífico Equatorial. É uma diferença padronizada da pressão medida no Tahiti e em Darwin, Australia. Períodos extensos de SOI negativo correspondem a períodos de pressão abaixo do normal no Tahiti e acima do normal em Darwin.Tais períodos conincidem com águas mais aquecidas no pacífico oriental, típicas do El Niño.
*   Season (2-Month):Intervalo de dois meses.
*   MEI.v2 - (bi-mmonthly Multivariate ENSO Index: A nova versão do índice MEI, o MEI.v2, é calculado com base em 5 variáveis relevantes para as condições atmosféricas e oceânica, são elas: pressão do nível do mar (sea level pressure -SLP), temperatura da superfície do mar ( sea surface temperature - SST), ventos zonais de superfície (surface zonal winds - U), ventos meridionais de superfície (surface meridional winds - V) e radiação de ondas longas emitidas (Outgoing Longwave Radiation - OLR). Ao integrar todas essas variáveis, o MEI trona-se um -indice mais estável e menos sucetivel a erros de instrmentaão, capaz de capturar com mais efici~encia a sazonalidade do ENOS. Valores positivos do MEI representam a fase quente da oscilação de temperatura - El Niño - enquanto valores negativos de MEI equivalem à fase mais fria - La niña. Durante o EL Niño as variáveis chaves que compões o índice MEI incluem SST anormalmente elevadono Pacífico equatorial, SLP anormalmente alto na indonésia e baixo na região mais próxia da América do Sul, redução  ou reversão no padrão dos ventos alísios, diminuição da convecção (OLR positivo) na Indonésia e Pacífico Oeste e aumento da mesma no Pacífico central (OLR negativo).
*   Season (3-Month): Intervalo de três meses.
*   ONI - Oceanic Niño Index: Este índice se tornou o padrão que a NOAA utiliza para classificar os fenômenos El Niño e La Niña no Oceano Pacifico Tropical. Compreende a média móvel trimestral da anomalia SST para a região Niño 3.4.Para ser classificado como  El Niño ou La Niña o índice deve ser maior que +0.5C ou menor que -0.5C, respectivamente, durente cinco meses consecutivos. (Definição operacional usada pelo NOAA).
*   Season (12-Month): intervalo de doze meses.
*   ENSO Phase-Intensity: Usando o índice ONI como base, a intensidade do ENOS pode ser subdividida em  WE= Weak El Niño (El Niño fraco), ME= Moderate El Niño (El Niño moderado), SE= Strong El Niño  (El Niño intenso), VSE= Very Strong El Niño (El Niño muito intenso), WL= Weak La Niña (La Niña fraco), ML= Moderate La Niña(La Niña moderado), SL= Strong La Niña (La Niña intenso).


Para ser caracterizado como um episódio de El Niño/La Niña as anomalias de temperaturas na superfície do mar (SST) devem seguir os critérios descritos abaixo por pelo menos 5 meses consecutivos:

**El Niño**

*   El Niño muito intenso -  Anomalia SST maior ou iagual a +2.0 C
*   El Niño intenso - Anomalia SST maior ou iagual a +1.5 a +1.9 C
*   El Niño moderado - Anomalia SST entre +1.0 e +1.4 C
*   El Niño fraco - Anomalia SST entre +0.5 e +0.9 C

**La Nniña**

*   La Niña intenso - Anomalia SST maior ou iagual a -1.5 C
*   La Niña moderado - Anomalia SST entre -1.0 e -1.4 C
*   La Niña fraco - Anomalia SST entre -0.5 e -0.9

Vale ressaltar que o perído neutro corresponde aos valores de SST entre -0.5 e +0.5 C

#3. Pré-Processamento de Dados

###3.1 Tratamento de Valores Nulos

Nesta etapa inicia-se a preparação dos dados, são tartados os valores nulos ou informações deixadas em branco,é realizada a limpeza dos dados e são tratadas as variáveis categóricas.

```python
#Criando um novo dataset para não sobrescrever o original

#Recuperando o nome das colunas
col = list(dataset.columns)

#Criando o novo dataset a partir de 1979, quando o índice MEI.v2 foi criado
atributos = dataset[(dataset['Year']>=1979)].reset_index(drop=True)


ms.matrix (atributos)
```

```python
#Substituindo NaN nas anomalias SST pela média

atributos['Nino 1+2 SST'].fillna(atributos['Nino 1+2 SST'].mean(), inplace = True)
atributos['Nino 1+2 SST Anomalies'].fillna(atributos['Nino 1+2 SST Anomalies'].mean(), inplace = True)
atributos['Nino 3 SST'].fillna(atributos['Nino 3 SST'].mean(), inplace = True)
atributos['Nino 3 SST Anomalies'].fillna(atributos['Nino 3 SST Anomalies'].mean(), inplace = True)
atributos['Nino 3.4 SST'].fillna(atributos['Nino 3.4 SST'].mean(), inplace = True)
atributos['Nino 3.4 SST Anomalies'].fillna(atributos['Nino 3.4 SST Anomalies'].mean(), inplace = True)
atributos['Nino 4 SST'].fillna(atributos['Nino 4 SST'].mean(), inplace = True)
atributos['Nino 4 SST Anomalies'].fillna(atributos['Nino 4 SST Anomalies'].mean(), inplace = True)

#Substituindo NaN de TNI pela Mediana

atributos['TNI'].fillna(atributos['TNI'].mean(), inplace = True)

#Substituindo os NaN de ONI, MEI.v2, SOI e PNA por 0
atributos['PNA'].fillna(0, inplace = True)
atributos['ONI'].fillna(0, inplace = True)
atributos['MEI.v2'].fillna(0, inplace = True)
atributos['SOI'].fillna(0, inplace = True)
atributos['OLR'].fillna(0, inplace = True)

#Removendo NaN de ENSO Phase-Intensity  deletando toda a linha
atributos.dropna(axis = 0, how = 'any', inplace = True)

#Exibindo visualização matricial da nulidade do dataset

ms.matrix(atributos)
```

###3.2 Tratamento das Variáveis categóricas

> One-Hot-Encoding


O One-Hot Encoding é uma técnica que transforma variáveis categóricas em representações binárias, criando colunas binárias para cada categoria. Essa abordagem é útil para algoritmos de aprendizado de máquina que requerem dados numéricos, melhorando a capacidade de modelagem de informações categóricas. No entanto, esta técnica cria um número adicional de colunas igual ao número de categorias únicas, aumentando a dimensionalidade dos dados. Isso pode ser problemático em conjuntos de dados grandes.

Neste trabalho o One-Hot Encoding será aplocado na coluna ENSO Phase-Intensity, que classifica os eventos de El Niño e La Niña em fracos, moderados, intensos ou muito intensos.


```python
# Convertendo a coluna para o tipo categórico
atributos['ENSO Phase-Intensity'] = atributos['ENSO Phase-Intensity'].astype('category')


# Criando os valores numericos e armazenando em outra coluna
atributos['ENSO Phase-Intensity_Codes'] = atributos['ENSO Phase-Intensity'].cat.codes


# Crinado o  One-hot-encoder
enc = OneHotEncoder(drop='first')

# Passing encoded columns
enc_Phase_Intensity = pd.DataFrame(enc.fit_transform(
    atributos[['ENSO Phase-Intensity_Codes']]).toarray())

# Adicionando as novas colunas ao dataframe original
atributos_encoded = atributos.join(enc_Phase_Intensity)

atributos_encoded.head()
```

```python
# Deletando variáveis categoricas que não serão usadas pelo algoritmo de ML
atributos_encoded = atributos_encoded.drop(['Date','Month','Season (2-Month)','Season (3-Month)','Season (12-Month)','ENSO Phase-Intensity'], axis = 1)

atributos_encoded.head()
```

# 4.Preparação dos Dados



####4.1 Separação dos dados

O metodo de separação dos dados usado será o Holdout. Nesta técnica os dados são separados em conjuntos distintos para o treinamento e a validação do modelo. Adicionalmente, será usado K-folds que particiona o dado em K-partes e em cada parte uma quantidade de daddos é usadas para treinamento e outra para validação.




```python
# Preparação dos dados

#Removendo a variável target dos atributos
atributos_machinelearning = atributos_encoded.drop(['MEI.v2'],axis=1)

# Separação em bases de treino e teste (holdout)
X = atributos_machinelearning.values
y = atributos_encoded['MEI.v2'].values # Separando a coluna de Target

# Convert continuous labels to discrete classes
y = [int(label) for label in y]

test_size = 0.30 # tamanho do conjunto de teste
seed = 4 # semente aleatória

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=test_size, shuffle=True, random_state=seed) # holdout

# Parâmetros e partições da validação cruzada
num_particoes = 5
kfold = KFold(n_splits=num_particoes, shuffle=True, random_state=seed)
```


####4.2 Seleção Atributos

A seleção de atributos é extremmente relevante para a melhora do desempenho dos modelos de *Machine Learning*. Os métodos de seleção de atributos avaliam relacionamento entre as variáveis de entrada e saída, para selecionar as variáveis de entrada que são mais relevantes ara o cálculo das variáveis de saída.

Utilizar todas todas as features não é considerado uma boa prática, pois atributos com  baixa importância podem estar sendo consideradas para o ajuste do modelo. Entre outros benefícios a seleção dos melhores atributos deixa o treinamento mais rápido, uma ves que reduz a dimensionalidade, melhora a precisão do modelo e reduz a chance de overfitting ou underfiting.

Aqui serão comparados dois metodos:


*   Select Kbest
*   Eliminação recursiva







> Select Kbest



```python
# SelectKBest

# Seleção de atributos com SelectKBest
best_var = SelectKBest(score_func=f_classif, k=8)

# Executa a função de pontuação em (X, y) e obtém os atributos selecionados
fit = best_var.fit(X_train, y_train)

# Reduz X para os atributos selecionados
features_kbest = fit.transform(X_train)

# Resultados
print('\nNúmero original de atributos:', X_train.shape[1])
print('\nNúmero reduzido de atributos:', features_kbest.shape[1])

# Exibe os atributos orginais
print("\nAtributos Originais:", atributos_machinelearning.columns[0:23])

# Exibe as pontuações de cada atributos e os 4 escolhidas (com as pontuações mais altas)
np.set_printoptions(precision=3) # 3 casas decimais
print("\nScores dos Atributos Originais:", fit.scores_)
print("\nAtributos Selecionados:", best_var.get_feature_names_out(input_features=atributos_machinelearning.columns[0:23]))
```



>  Eliminação recursiva



```python
# Eliminação Recursiva de Atributos

# Criação do modelo
lr= DecisionTreeRegressor()

# Eliminação Recursiva de Atributos
rfe=RFE(estimator=lr, n_features_to_select=8)

features_rfe=rfe.fit_transform(X_train, y_train)


# Print dos resultados
print("Atributos Originais:", atributos_machinelearning.columns[0:23])

# Exibe os atributos selecionados (marcados como True em "Atributos Selecionados"
# e com valor 1 em "Ranking dos Atributos")
print("\nAtributos Selecionados: %s" % rfe.support_)
print("\nRanking de atributos: %s" % rfe.ranking_)
print("\nQtd de melhores Atributos: %d" % rfe.n_features_)
print("\nNomes dos Atributos Selecionados: %s" % rfe.get_feature_names_out(input_features=atributos_machinelearning.columns[0:23]))
```

Os resultados foram distintos, portanto os modelos serão testados e comparados com as duas seleções de atributos.

#5. Modelagem e treinamento

Nesta seção serão treinados e comparados os resultados de vários tipos de modelos de regressão de Machine learning. A ferramente utilizada para avaliar o desempenho do modelo será o cross-validation score. Essa função executa a validação cruzada de acordo com o número de K-partições informada no parâmetro cv. Em  seguida, ela retorna uma métrica de performance para cada K-partições de teste.

As metricas escolhidas fora as seguintes
 Raiz Quadrada do Erro Quadrático Médio — ou simplesmente RMSE

*   RMSE (Raiz Quadrada do Erro Quadrático Médio) — é a raiz quadrada da média dos erros ao quadrado, na prática, é diferença entre o valor que foi previsto pelo seu modelo e o valor real que foi observado. Quanto menor o valor do RMSE, melhor o desempenho do modelo.
*   MSE (Erro Quadrático Médio) - corresponde à média dos erros ao quadrado.



###5.1 Modelagem Utilizando Todos os Atributos

```python
# Modelagem utilizando todos os atributos

# Definindo uma seed global para esta célula de código
np.random.seed(7)

# Listas para armazenar os modelos, os resultados e os nomes dos modelos
models = []
results = []
names = []

# Preparando os modelos e adicionando-os em uma lista
models.append(('LR', LinearRegression()))
models.append(('Ridge', Ridge()))
models.append(('Lasso', Lasso()))
models.append(('KNN', KNeighborsRegressor()))
models.append(('CART', DecisionTreeRegressor()))
models.append(('RFOR',RandomForestRegressor()))
models.append(('SVM', SVR()))
models.append(('Xbot',XGBRegressor()))

# Avaliando um modelo por vez
for name, model in models:
  cv_results = cross_val_score(model, X_train, y_train, cv=kfold, scoring='neg_mean_squared_error')
  results.append(cv_results)
  names.append(name)
  # imprime MSE, desvio padrão do MSE e RMSE dos 10 resultados da validação cruzada
  msg = "%s: MSE %0.2f (%0.2f) - RMSE %0.2f" % (name, abs(cv_results.mean()), cv_results.std(), np.sqrt(abs(cv_results.mean())))
  print(msg)

# Boxplot de comparação dos modelos
fig = plt.figure()
fig.suptitle('Comparação do MSE dos Modelos')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```

Para este conjunto de dados o modelo com o melhor desempenho foi o Random Forest. No entanto há outros com o desempenho próximo.
Com o intuito de tentar melhorar o desempenho dos modelos, será realizado o ajuste de hiperparâmetros nos dois modelos de melhor desempenho e também no de pior desempenho.
